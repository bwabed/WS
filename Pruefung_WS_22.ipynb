## Aufgabe 4
Je sechs Äpfel der Sorte Idared werden in eine Obstschale abgepackt. Das Gewicht der Äpfel dieser Sorte ist in einer bestimmten Saison normalverteilt, mit einem Mittelwert von 190 g und einer Varianz von 200 g2. Für die Obstschale dürfen nur Äpfel mit einem Gewicht von mindestens 170 g verwendet werden.

a) Welcher prozentuale Anteil der Ernte kann nicht in Obstschalen abgepackt werden? (3 Pkt.)

b) Für Premium-Obstschalen mit lauter Idared-Äpfeln soll ein Anteil von 20% aller Äpfel dieser Sorte verwendet werden, symmetrisch um das Gewicht 190 g verteilt. Wie muss dazu die untere bzw. obere Grenze des Gewichts für die Äpfel festgelegt werden (auf Gramm genau)? (3 Pkt.)

Um die Fragen zu beantworten, verwenden wir die Eigenschaften der Normalverteilung. Die Normalverteilung wird durch zwei Parameter beschrieben: den Mittelwert ($\mu$) und die Standardabweichung ($\sigma$). Die Varianz ist das Quadrat der Standardabweichung ($\sigma^2$).

Gegeben sind:
- Mittelwert ($\mu$) = 190 g
- Varianz ($\mu^2$) = 200 g²
- Standardabweichung ($\sigma$) = √200 g² = √200 g

### Teil a)

Wir müssen den prozentualen Anteil der Äpfel berechnen, die weniger als 170 g wiegen, um den Anteil zu finden, der nicht in die Obstschalen abgepackt werden darf.

Dazu verwenden wir die Z-Score-Formel, um den Wert von 170 g in einen Z-Score umzurechnen:
$ Z = \frac{X - \mu}{\sigma} $

wobei $X$ das Gewicht des Apfels ist. Nachdem wir den Z-Score berechnet haben, können wir eine Z-Tabelle (Standard-Normalverteilungstabelle) oder eine kumulative Verteilungsfunktion (CDF) verwenden, um den prozentualen Anteil der Äpfel zu bestimmen, der weniger als 170 g wiegt.

### Teil b)

Um die untere und obere Grenze für die Premium-Obstschalen zu bestimmen, bei denen genau 20% der Äpfel verwendet werden, müssen wir die Grenzen so festlegen, dass 10% der Äpfel leichter als die untere Grenze und 10% schwerer als die obere Grenze sind, da die Verteilung symmetrisch um den Mittelwert von 190 g ist.

Wir verwenden die Z-Tabelle rückwärts, um die Z-Scores zu finden, die diesen Prozentsätzen entsprechen, und wandeln diese dann in die entsprechenden Gewichtsgrenzen um.

Lassen Sie uns zuerst den prozentualen Anteil der Ernte berechnen, der nicht in Obstschalen abgepackt werden kann (Teil a), und dann die Grenzen für die Premium-Obstschalen bestimmen (Teil b).

### Ergebnisse

#### Teil a)

Der prozentuale Anteil der Ernte, der nicht in Obstschalen abgepackt werden kann, weil die Äpfel weniger als 170 g wiegen, beträgt etwa 7.87%. Das bedeutet, dass ungefähr 7.87% der Äpfel für die Standard-Obstschalen nicht verwendet werden können.

#### Teil b)

Für die Premium-Obstschalen, bei denen ein Anteil von 20% aller Äpfel dieser Sorte verwendet werden soll, müssen die Gewichtsgrenzen wie folgt festgelegt werden:
- Die untere Gewichtsgrenze muss bei etwa 171.88 g festgelegt werden.
- Die obere Gewichtsgrenze muss bei etwa 208.12 g festgelegt werden.

Diese Grenzen stellen sicher, dass die Äpfel, die in die Premium-Obstschalen kommen, symmetrisch um das mittlere Gewicht von 190 g verteilt sind, mit jeweils 10% der leichteren und 10% der schwereren Äpfel außerhalb dieser Grenzen.

```python
# Gegebene Werte
mu = 190  # Mittelwert in g
varianz = 200  # Varianz in g²
sigma = sqrt(varianz)  # Standardabweichung in g
grenze = 170  # Grenzgewicht in g

# Berechnung des prozentualen Anteils, der weniger als 170 g wiegt
z_score_untergrenze = (grenze - mu) / sigma
anteil_unter_170g = pnorm(z_score_untergrenze)

# Ausgabe
anteil_unter_170g

x = rnorm(100, sd=sigma,mean=mu)
hist(x,breaks=200)
```

```python
# Berechnung der Z-Scores für die unteren 10% und oberen 10%
z_score_untere_10_prozent = qnorm(0.4)
z_score_obere_10_prozent = qnorm(0.6)

# Umrechnung der Z-Scores in Gewichtsgrenzen
untere_grenze_gewicht = mu + (z_score_untere_10_prozent * sigma)
obere_grenze_gewicht = mu + (z_score_obere_10_prozent * sigma)

# Ausgabe
untere_grenze_gewicht
obere_grenze_gewicht
```

## Aufgabe 5
Ein Arzt schätzt, dass die Standardabweichung bei der Messung der Reaktionszeit 0,05 Sekunden (s) beträgt. Wie gross müssen die Stichproben seiner Messungen sein um

a) zu 95% sicher zu sein, dass sein Schätzfehler 0,01 s nicht überschreitet? (3 Pkt.)

b) zu 99% sicher zu sein, dass sein Schätzfehler 0,01 s nicht überschreitet? (3 Pkt.)


Um die Größe der Stichproben zu bestimmen, die benötigt wird, um mit einer bestimmten Konfidenz sicherzustellen, dass der Schätzfehler einen bestimmten Wert nicht überschreitet, kann die Formel für den Stichprobenumfang bei Schätzungen der Mittelwerte verwendet werden. Diese Formel lautet:

$ n = \left(\frac{Z \cdot \sigma}{E}\right)^2 $

Dabei ist:
- $n$ die erforderliche Stichprobengröße,
- $Z$ der Z-Wert, der dem gewünschten Konfidenzniveau entspricht (dieser Wert kommt aus der Standard-Normalverteilung und hängt vom gewählten Konfidenzniveau ab),
- $\sigma$ die geschätzte Standardabweichung der Population (in diesem Fall die Messung der Reaktionszeit),
- $E$ der maximal akzeptable Schätzfehler (Fehlergrenze).

Für die Konfidenzniveaus von 95% und 99% müssen wir die entsprechenden Z-Werte finden. Typischerweise verwendet man für ein Konfidenzniveau von 95% einen Z-Wert von etwa 1,96 und für 99% einen Z-Wert von etwa 2,576.

Setzen wir die gegebenen Werte ein, um die erforderlichen Stichprobengrößen zu berechnen:

- $\sigma = 0,05$ s (Standardabweichung),
- $E = 0,01$ s (maximal akzeptierter Schätzfehler).

### a) Für ein Konfidenzniveau von 95%

$ n = \left(\frac{1,96 \cdot 0,05}{0,01}\right)^2 $

### b) Für ein Konfidenzniveau von 99%

$ n = \left(\frac{2,576 \cdot 0,05}{0,01}\right)^2 $

Berechnen wir nun die erforderlichen Stichprobengrößen für beide Konfidenzniveaus.

### Ergebnisse

#### a) Für ein Konfidenzniveau von 95%

Die erforderliche Stichprobengröße, um mit 95% Sicherheit zu sein, dass der Schätzfehler 0,01 s nicht überschreitet, beträgt etwa 96 Messungen.

#### b) Für ein Konfidenzniveau von 99%

Die erforderliche Stichprobengröße, um mit 99% Sicherheit zu sein, dass der Schätzfehler 0,01 s nicht überschreitet, beträgt etwa 166 Messungen.

Diese Berechnungen zeigen, wie die Größe der Stichprobe erhöht werden muss, um eine höhere Genauigkeit (niedrigeren Schätzfehler) bei höherem Konfidenzniveau zu gewährleisten.

```python
# Gegebene Werte
sigma <- 0.05  # Standardabweichung in s
E <- 0.01  # Maximal akzeptierter Schätzfehler in s

# Z-Werte für die Konfidenzniveaus
Z_95 <- 1.96
Z_99 <- 2.576

# Berechnung der Stichprobengrößen
n_95 <- (Z_95 * sigma / E)^2
n_99 <- (Z_99 * sigma / E)^2

# Ausgabe der Stichprobengrößen
cat("Stichprobengröße für 95% Konfidenzniveau:", round(n_95), "\n")
cat("Stichprobengröße für 99% Konfidenzniveau:", round(n_99), "\n")
```
